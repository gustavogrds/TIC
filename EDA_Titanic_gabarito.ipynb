```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic_url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_data = pd.read_csv(titanic_url)
```

```python
# Questão 1 : Descreva o dataset
dataset_description = titanic_data.describe()
```

```python
# Questào 2; encontre o passageiro mais jovem e o mais idoso
oldest_passenger = titanic_data[titanic_data['Age'] ==titanic_data['Age'].max()]
print(oldest_passenger)
youngest_passenger = titanic_data[titanic_data['Age'] == titanic_data['Age'].min()]
print(youngest_passenger)
```

```python
# Questão 3: Faça um gráfico da distribuição de idades dos passageiros
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x='Age', bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Questão 4: Como os passageiros estão distribuídos por classe? Grafique.
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_data, x='Pclass')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
(titanic_data[titanic_data['Survived']==1]['Sex'].value_counts())/(len(titanic_data))
```

```python
# Questão 5: Calcule a taxa de sobreviventes por Sexo
gender_survival_rate = titanic_data.groupby('Sex')['Survived'].mean()
gender_survival_rate
```

```python
# Questão 6: Calcule a tarifa média pela Classe (PClass) do ticket
average_fare_by_class = titanic_data.groupby('Pclass')['Fare'].std()
average_fare_by_class

titanic_data.Pclass.mean()
titanic_data['Pclass'].mean()
```

```python
# Questão 7: Plote a taxa de sobrevivência pela classe (PClass) de passageiros
plt.figure(figsize=(8, 6))
sns.barplot(data=titanic_data, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Questão 8: Substitua as idades ausentes pela idade mediana e calcule a nova idade média
median_age = titanic_data['Age'].median()
titanic_data['Age'].fillna(median_age, inplace=True)
new_average_age = titanic_data['Age'].mean()
new_average_age
```

```python
# Questão 9: Substitua os valores "masculino" e "feminino" na coluna "Sexo" por "M" e "F"
titanic_data['Sex'].replace({'male': 'M', 'female': 'F'}, inplace=True)
titanic_data['Sex']
```

```python
# Pergunta 10: Selecione os passageiros que estavam na primeira classe e pagaram uma tarifa superior a US$ 100
high_fare_first_class = titanic_data[(titanic_data['Pclass'] == 1) & (titanic_data['Fare'] > 100)]
high_fare_first_class
```
